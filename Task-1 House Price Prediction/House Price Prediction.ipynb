# Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#  Reading Data

```python
HouseDT = pd.read_csv('Housing.csv')
HouseDT.head()
```

```python
HouseDT.isnull().sum()
```

```python
HouseDT.fillna(method='bfill')
```

```python
HouseDT.dropna(axis=1)
```

```python
HouseDT.info()
```

```python
HouseDT.describe()
```

```python
X = HouseDT[['Avg. Area Income','Avg. Area Number of Bedrooms', 'Avg. Area Number of Rooms', 'Avg. Area House Age']]

y = HouseDT['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4) 
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mo=LinearRegression()
```

```python
mo.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_pred = mo.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"MSE: {mse:.2f}, R-squared: {r2:.2f}")
```
